```python
import cv2 as cv
import numpy as np
import os
```

```python
files = ["panorma1.jpg","panorma2.jpg","panorma3.jpg","panorma4.jpg","panorma5.jpg","panorma6.jpg"]
for i in files:
    path = os.path.join("data", i)
    print(path)
    img = cv.imread(r"{}".format(path))
    cv.imshow("img", img)
    cv.waitKey(1000)
    cv.destroyWindow("img")
```

```python
files = ["panorma1.jpg","panorma2.jpg","panorma3.jpg","panorma4.jpg","panorma5.jpg","panorma6.jpg"]
pathes = []
image_coordinations = {}
for i in files:
    path = os.path.join("data", i)
    pathes.append(path)
    img = cv.imread(r"{}".format(path))

    divided_by = img.shape[1] // 4
    start_index = 0
    end_index = divided_by
    indexes = []
    for j in range(4):
        indexes.append([start_index, end_index-1])
        start_index = end_index
        end_index += divided_by

        
    image_coordinations[i[:-4]] = indexes.copy()
    indexes.clear()
    


print(image_coordinations)


for inx, adrs in list(zip(image_coordinations, pathes)):
    img = cv.imread(r"{}".format(adrs))
    for j in image_coordinations[inx]:
        seprated = img[:, j[0]:j[1]]
        cv.imshow("seprated", seprated)
        cv.waitKey(800)
        cv.destroyWindow("seprated")
        
        

```
