```python
files = ["panorma1.jpg","panorma2.jpg","panorma3.jpg","panorma4.jpg","panorma5.jpg","panorma6.jpg", "panorma7.jpg"]
pathes = []
image_coordinations = {}
for i in files:
    path = os.path.join("data", i)
    pathes.append(path)
    img = cv.imread(r"{}".format(path))

    divided_by = img.shape[1] // 2
    start_index = 0
    end_index = divided_by
    indexes = []
    for j in range(2):
        if j == 0:
            indexes.append([start_index, end_index+70])
        elif j == 1:
            indexes.append([start_index-70, end_index])
        else:
            indexes.append([start_index-20, end_index+20])

        start_index = end_index
        end_index += divided_by

        
    image_coordinations[i[:-4]] = indexes.copy()
    indexes.clear()
    


print(image_coordinations)

images = []
one_img = []
for inx, adrs in list(zip(image_coordinations, pathes)):
    img = cv.imread(r"{}".format(adrs))

    for j in image_coordinations[inx]:
        seprated = img[:, j[0]:j[1]]
        cv.imshow("seprated", seprated)
        cv.waitKey(900)
        cv.destroyWindow("seprated")
        one_img.append(seprated.copy())

    images.append(one_img.copy())
    one_img.clear()        

```

```python
for i, img in enumerate(images):
    section1 = img[0]
    section2 = img[1]
    # section3 = i[2]
    # section4 = i[3]

    stitcher = cv.Stitcher_create()
    status, stitched_image = stitcher.stitch((section1, section2))

    if status == cv.Stitcher_OK:
        cv.imshow("stitched_image",stitched_image)
        cv.waitKey()
        cv.destroyWindow("stitched_image")
    elif status == cv.Stitcher_ERR_NEED_MORE_IMGS:
        print("Stitcher_ERR_NEED_MORE_IMGS")
    elif status == cv.Stitcher_ERR_CAMERA_PARAMS_ADJUST_FAIL:
        print("Stitcher_ERR_CAMERA_PARAMS_ADJUST_FAIL")
    elif status == cv.Stitcher_ERR_HOMOGRAPHY_EST_FAIL:
        print("Stitcher_ERR_HOMOGRAPHY_EST_FAIL")

    
    
    
```
