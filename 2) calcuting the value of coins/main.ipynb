```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread(r"data\coin1.jpg")
plt.imshow(img[...,::-1])
```

```python
img = cv.imread(r"data\coin1.jpg")

gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

gray = cv.GaussianBlur(gray, (3,3), 1)


circles = cv.HoughCircles(image=gray, method=cv.HOUGH_GRADIENT, dp=1.2, minDist=100, param1=500, param2=30, minRadius=1, maxRadius=150)


if circles is not None:
    circles = np.round(circles[0, :]).astype(int)
    for x,y,r in circles:
        cv.circle(img, (x,y), r, (0,255,0), 2)
        cv.putText(img, str(r), (x,y), cv.FONT_HERSHEY_COMPLEX, 1, (0,0,255), 1)


plt.imshow(img[...,::-1])
```

```python
ListOfPrices = []
for x, y, r in circles:
    if r > 50 and r < 56:
        ListOfPrices.append(50)
    elif r > 59 and r < 63:
        ListOfPrices.append(100)
    elif r > 63 and r < 66:
        ListOfPrices.append(250)
    elif r > 67 and r < 73:
        ListOfPrices.append(500)

sum_price = sum(ListOfPrices)
print(f"sum: {sum_price}")
```

```python
cv.putText(img, str(sum_price), (10,30), cv.FONT_HERSHEY_COMPLEX, 1, (0,255,0), 1)
plt.imshow(img[...,::-1])
```
